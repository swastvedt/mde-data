```python
###Load total student enrollment data from 'raceEthnicity' file
enrolldf = pd.read_csv('data/' + 'raceEthnicity.csv',
    dtype = {0:str, 1:str, 2:str, 3:str},
    usecols = ['DataYear', 'DistrictID', 'Grade', 'DistrictName', 'TotalStudents'],
    index_col = [0,1,2]
)
```

```python
###Select only All Grades, SPPS enrollment
FLEnroll = enrolldf.xs(('0831-01-000', 'All Grades'), level=[1, 2])

###Select open enroll in data for SPPS, summed by year
FLIn = data.xs(('0831-01-000'), level=1)['CountofOpenEnrolled'].groupby(by='DataYear', level=0).sum()
```

```python
###Join enrollment in, out, and total enrollment
FL = FLEnroll.join(FLIn.to_frame(), how='outer').join(FLOut.to_frame(), how='outer', rsuffix='O', lsuffix='I')
```

```python
###Convert open enrollment columns to decimal numbers
FL['CountofOpenEnrolledI'] = pd.to_numeric(FL['CountofOpenEnrolledI'])
FL['CountofOpenEnrolledO'] = pd.to_numeric(FL['CountofOpenEnrolledO'])

###Calculate net loss: enrolled in - enrolled out
FL['NetLoss'] = FL['CountofOpenEnrolledI'] - FL['CountofOpenEnrolledO']
```

```python
###Convert net loss to a percent of total enrollment for the given year
FL['NetLossPercent'] = FL['NetLoss']/FL['TotalStudents']*100
FL
```

```python
###Select open enroll in data, summed by enrolling district
enrollIn = data['CountofOpenEnrolled'].groupby(by=['DataYear', 'EnrollDistrictID'], level=[0,1]).sum()
enrollIn.head()
```

```python
###Select open enroll out data, summed by resident district
enrollOut = data['CountofOpenEnrolled'].groupby(by=['DataYear', 'ResidentDistrictID'], level=[0,2]).sum()
enrollOut.head()
```

```python
###Join enrollment in, out, and total enrollment
openEnroll = enrollment.reset_index().merge(
    enrollIn.to_frame().reset_index(),
    left_on=['DataYear', 'DistrictID'],
    right_on=['DataYear', 'EnrollDistrictID']
).merge(
    enrollOut.to_frame().reset_index(),
    left_on=['DataYear', 'DistrictID'],
    right_on=['DataYear', 'ResidentDistrictID'],
    suffixes=['I', 'O']
).set_index(
    ['DataYear', 'DistrictID']
).drop(['EnrollDistrictID', 'ResidentDistrictID'], axis=1)
```

```python
###Convert open enrollment columns to decimal numbers
openEnroll['CountofOpenEnrolledI'] = pd.to_numeric(openEnroll['CountofOpenEnrolledI'])
openEnroll['CountofOpenEnrolledO'] = pd.to_numeric(openEnroll['CountofOpenEnrolledO'])

###Calculate net loss: enrolled in - enrolled out
openEnroll['NetLoss'] = openEnroll['CountofOpenEnrolledO'] - openEnroll['CountofOpenEnrolledI']
```

```python
###Convert net loss to a percent of total enrollment for the given year
openEnroll['NetLossPercent'] = openEnroll['NetLoss']/openEnroll['TotalStudents']*100
```

```python
openEnroll.xs('16-17', level=0).sort_values(by='NetLossPercent', ascending=False)
```
